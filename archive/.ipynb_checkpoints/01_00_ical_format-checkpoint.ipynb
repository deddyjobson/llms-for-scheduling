```python
import pandas as pd
from icalendar import Calendar

# Load the data from the ICS file
with open('event-calendar-2022.ics', 'r') as f:
    cal = Calendar.from_ical(f.read())
```

```python
events_data = []
# Iterate over each event in the calendar
for event in cal.walk('vevent'):
    # Extract relevant information
    try:
        title, authors = event.get('summary').split(" - ", 1)  # Expecting exactly 2 items
        authors = authors.split('\,')
    except ValueError:  # If not exactly 2 items, then it's a ValueError
        title = event.get('summary')  # Take the whole summary as title
        authors = None  # Default, no authors found

    event_data = {
        'date': event.get('dtstart').dt.date(),
        'start_time': event.get('dtstart').dt.time(),
        'duration': event.get('dtend').dt - event.get('dtstart').dt,
        'title': title,
        'authors': authors,
    }
    events_data.append(event_data)
```

```python
# Create a pandas DataFrame
events_df = pd.DataFrame(events_data)

# Print the DataFrame
events_df
```
